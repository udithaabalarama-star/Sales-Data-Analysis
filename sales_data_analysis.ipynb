# Sales Data Analysis using Python

## Objective

The goal of this project is to analyze sales data to understand trends, product performance, and regional sales distribution.

## Dataset Description

The dataset contains information about:
- Date of sale
- Product sold
- Sales amount
- Quantity sold
- Region

```python
import pandas as pd

data = {
    "Date": [
        "2024-01-01","2024-01-02","2024-01-03","2024-01-04",
        "2024-01-05","2024-01-06","2024-01-07","2024-01-08"
    ],
    "Product": ["Laptop","Phone","Tablet","Laptop","Phone","Tablet","Laptop","Phone"],
    "Region": ["South","North","East","West","South","East","North","West"],
    "Sales": [50000,30000,20000,55000,32000,22000,60000,35000],
    "Quantity": [5,10,7,6,11,8,7,12]
}

df = pd.DataFrame(data)
df
```

## Data Processing

- Converted Date column into datetime format
- Extracted month from date
- Prepared data for analysis

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Month"] = df["Date"].dt.month
```

```python
df[["Sales","Quantity"]].corr()
```

## Data Analysis & Visualization

```python
import matplotlib.pyplot as plt

df.plot(x="Date", y="Sales", kind="line", marker="o")
plt.title("Sales Trend Over Time")
plt.show()
```

```python
# Total sales
print("Total Sales:", df["Sales"].sum())

# Average sales
print("Average Sales:", df["Sales"].mean())
```

```python
# Sales by product
df.groupby("Product")["Sales"].sum()
```

```python
import matplotlib.pyplot as plt

df.groupby("Product")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Product")
plt.show()
```

```python
# Sales by Region
df.groupby("Region")["Sales"].sum()
```

```python
df.groupby("Product")["Quantity"].sum()
```

```python
df.groupby("Region")["Sales"].sum().plot(kind="pie", autopct='%1.1f%%')
plt.title("Sales Distribution by Region")
plt.show()
```

## Key Insights

- Laptop generates the highest revenue among all products
- Phone shows consistent sales over time
- South and West regions contribute significantly to overall sales
- Product demand varies across regions, indicating regional preferences
